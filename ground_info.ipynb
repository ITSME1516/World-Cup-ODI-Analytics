```python
import pandas as pd 
import bs4
import requests
import csv
```

```python
data = pd.read_csv("./Data/all_match_info.csv")
data = data[["Team 1","Team 2","Scorecard","Ground","Ground Link"]]
```

```python
data
```

```python
match_id = data.Scorecard
ground = data["Ground Link"]
dic_data = []
for id, ground in zip(match_id,ground):
    code = requests.get(ground)
    soup = bs4.BeautifulSoup(code.content,"html.parser")
    card = soup.find("section", attrs={"class":"ds-bg-raw-black ds-p-4"})
    card = card.find_all("span")
    data = {}
    data["id"] = id
    #data["Ground"] = ground
    data["location"] = card[1].text
    tabel = soup.find("table")
    for rows in tabel.find_all("tr"):
        data[list(rows.children)[0].text] = list(rows.children)[1].text
    
    dic_data.append(data)
```

```python
ground_data = pd.DataFrame(dic_data)
ground_data.to_csv("./Data/Ground Data.csv",index=False)
```
