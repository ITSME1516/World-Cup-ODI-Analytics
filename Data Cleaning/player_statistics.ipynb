# Batting Statistics:
Runs Scored: Total runs scored by the player.

Batting Average: Runs scored per innings.

Strike Rate: Runs scored per 100 balls faced.

Centuries: Total centuries (100+ runs) scored.

Half-Centuries: Total half-centuries (50+ runs) scored.

Highest Score: The highest score in an innings.

Balls Faced: Total number of balls faced.

Boundary Count: Number of fours and sixes hit.

Bowling Average: Runs given per wicket taken.

Economy Rate: Runs conceded per over bowled.

Strike Rate: Balls bowled per wicket taken.

Five-Wicket Hauls: Number of instances of taking five or more wickets in an innings.

Best Bowling Figures: Best performance in a single innings.

Overs Bowled: Total overs bowled.

Maidens: Total maiden overs bowled (overs with no runs conceded).

# Fielding Statistics:
Catches: Total catches taken.

Stumpings: Total stumpings made (for wicketkeepers).

Run-outs: Total run-outs effected.

# All-Rounder Statistics:
Batting and Bowling Average: Combined impact of batting and bowling.

Player Impact Score: A metric to measure the overall impact of a player in matches.

# Additional Factors:
Match Performance: Performance in crucial matches or against specific teams.

Player Form: Recent form and consistency over a period.

Injury History: History of injuries and fitness levels.

Captaincy Record: For players who have captained the team.

Experience: Number of matches played, World Cup appearances.

# Batting Statistics:
Runs Scored: Total runs scored by the player.

Batting Average: Runs scored per innings.

Strike Rate: Runs scored per 100 balls faced.

Centuries: Total centuries (100+ runs) scored.

Half-Centuries: Total half-centuries (50+ runs) scored.

Highest Score: The highest score in an innings.

Balls Faced: Total number of balls faced.

```python
# import modules
import pandas as pd
import numpy as np 
from scipy import stats
import math
```

```python
# importing Data

batting_data = pd.read_csv("./Data/pre_batting_data.csv")
batting_data.info()
```

```python
#Runs Scored: Total runs scored by the player.

print("Total Runs Scored by Each Player".center(50,"-"))
total_runs = batting_data.groupby("Player")["Runs"].sum()
total_runs = pd.DataFrame(total_runs).reset_index()
total_runs.Runs = total_runs.Runs.apply(lambda x: round(x,2))
print(total_runs.head())

#Batting Average: Runs scored per innings.

print("Average Runs Scored by Each Player".center(50,"-"))
average_runs = batting_data.groupby("Player")["Runs"].mean()
average_runs = pd.DataFrame(average_runs).reset_index()
average_runs.Runs = average_runs.Runs.apply(lambda x: round(x,2))
average_runs = average_runs.rename(columns={"Runs":"Average Runs"})
print(average_runs.head())

#Strike Rate: Runs scored per 100 balls faced.

print("Strike Rate by Each Player".center(50,"-"))
strike_rate = batting_data.groupby("Player")["Strike Rate"].mean()
strike_rate = pd.DataFrame(strike_rate).reset_index()
strike_rate["Strike Rate"] = strike_rate["Strike Rate"].apply(lambda x: round(x,2))
print(strike_rate.head())
```

```python
#Centuries: Total centuries (100+ runs) scored.

print("Total centuries (100+ runs) scored by Each Player".center(50,"-"))
batting_data["Centuries"] = batting_data["Runs"].apply(lambda x:1 if x>=100 else 0)

total_centuries = batting_data[["Player","Centuries"]]
total_centuries = pd.DataFrame(total_centuries.groupby("Player")["Centuries"].sum()).reset_index()
print(total_centuries.head(),end='\n\n\n')

#Half-Centuries: Total half-centuries (50+ runs) scored.

print("Total half centuries (50+ runs) scored by Each Player".center(50,"-"))
batting_data["Half Centuries"] = batting_data["Runs"].apply(lambda x:1 if x>=50 else 0)

total_Half_centuries = batting_data[["Player","Half Centuries"]]
total_Half_centuries = pd.DataFrame(total_Half_centuries.groupby("Player")["Half Centuries"].sum()).reset_index()
print(total_Half_centuries.head(),end='\n\n\n')

#Highest Score: The highest score in an innings.

print("The highest score in an innings".center(50,"-"))

highest_score = batting_data[["Player","Runs"]]
highest_score = pd.DataFrame(highest_score.groupby("Player")["Runs"].max()).reset_index()
highest_score = highest_score.rename(columns={"Runs":"Highest Run"})
print(highest_score.head(),end='\n\n\n')

```

```python
#Balls Faced: Total number of balls faced.

print("Total number of balls faced by Each Player".center(50,"-"))

no_balls = batting_data[["Player","Balls"]]
no_balls = pd.DataFrame(no_balls.groupby("Player")["Balls"].sum()).reset_index()
no_balls = no_balls.rename(columns={"Balls":"No. of Balls"})
print(no_balls.head(),end='\n\n\n')

#Average Balls Faced: Average number of balls faced.

print("Average number of balls faced by Each Player".center(50,"-"))

avg_balls = batting_data[["Player","Balls"]]
avg_balls = pd.DataFrame(avg_balls.groupby("Player")["Balls"].mean()).reset_index()
avg_balls["Balls"] = avg_balls["Balls"].apply(lambda x : math.ceil(x))
avg_balls = avg_balls.rename(columns={"Balls":"Average Balls Faced"})
print(avg_balls.head(),end='\n\n\n')
```

```python
batting_stats = total_runs.merge(average_runs).merge(strike_rate).merge(total_centuries).merge(total_Half_centuries).merge(highest_score).merge(no_balls).merge(avg_balls)
```

```python
batting_stats.head()
```

```python
batting_stats.to_csv("./Data/Batting Statistics.csv",index=False)
```

# Bowling Statistics:
Wickets Taken: Total wickets taken by the player.

Bowling Average: Runs given per wicket taken.

Economy Rate: Runs conceded per over bowled.

Strike Rate: Balls bowled per wicket taken.

Five-Wicket Hauls: Number of instances of taking five or more wickets in an innings.

Best Bowling Figures: Best performance in a single innings.

Overs Bowled: Total overs bowled.

Maidens: Total maiden overs bowled (overs with no runs conceded).

```python
bowling_data = pd.read_csv("./Data/pre_bowling_data.csv")
bowling_data.info()
```

```python
#Wickets Taken: Total wickets taken by the player.


print("Total wickets taken by the player".center(50,"-"))

total_wicket = bowling_data[["Player","Wickets"]]
total_wicket = pd.DataFrame(total_wicket.groupby("Player")["Wickets"].sum()).reset_index()
total_wicket = total_wicket.rename(columns={"Wickets":"No. of Wickets"})
print(total_wicket.head(),end='\n\n\n')

#Bowling Average: Runs given per wicket taken.

print("Runs given per wicket taken".center(50,"-"))
bowling_data["Bowling Average"] = bowling_data.apply(lambda x: x["Given Runs"]/x["Wickets"] if x["Wickets"] !=0 else 0 ,axis=1)
bowling_data["Bowling Average"] = bowling_data["Bowling Average"].apply(lambda x: round(x,2))
bowling_average = bowling_data[["Player","Bowling Average"]]
bowling_average = pd.DataFrame(bowling_average.groupby("Player")["Bowling Average"].mean())
bowling_average = bowling_average.reset_index()
bowling_average["Bowling Average"] = bowling_average["Bowling Average"].apply(lambda x: round(x,2))
print(bowling_average.head(),end='\n\n\n')

#Average Economy Rate: Runs conceded per over bowled.

print("Runs conceded per over bowled".center(50,"-"))
economy_rate = bowling_data[["Player","ECON"]]
economy_rate = pd.DataFrame(economy_rate.groupby("Player")["ECON"].mean())
economy_rate = economy_rate.reset_index()
economy_rate["ECON"] = economy_rate["ECON"].apply(lambda x: round(x,2))
economy_rate = economy_rate.rename(columns={"ECON":"Average ECON"})
print(economy_rate.head())
```

```python
#Total Balls Bowled: Number of balls bowled in inning.

bowling_data["Balls"] = bowling_data.apply(lambda x: (int(str(x["Overs"]).split(".")[0])*6) + int(str(x["Overs"]).split(".")[1]) ,axis=1)

#Average Strike Rate: Balls bowled per wicket taken.

bowling_data["Strike Rate"] = bowling_data.apply(lambda x: round(x["Balls"]/x['Wickets'],2) if x['Wickets']!=0 else 0,axis=1)
print("Balls bowled per wicket taken".center(50,"-"))
average_strike_rate = pd.DataFrame(bowling_data.groupby("Player")["Strike Rate"].mean()).reset_index()
average_strike_rate = average_strike_rate.rename(columns={"Strike Rate":"Average Strike Rate"})
average_strike_rate["Average Strike Rate"] = average_strike_rate["Average Strike Rate"].apply(lambda x:round(x,2))
print(average_strike_rate.head(),end="\n\n\n")

#Five-Wicket Hauls: Number of instances of taking five or more wickets in an innings.

print("Number of instances of taking five or more wickets in an innings".center(50,"-"))
bowling_data["Five Wicket Hauls"] = bowling_data.apply(lambda x:1 if x["Wickets"]>=5 else 0,axis=1)
five_wicket = pd.DataFrame(bowling_data.groupby("Player")["Five Wicket Hauls"].sum()).reset_index()
five_wicket = five_wicket.rename(columns={"Five Wicket Hauls":"Total Five Wicket Hauls"})
print(five_wicket.head(),end="\n\n\n")

#Best Bowling Figures: Best performance in a single innings.
print("Best performance in a single innings".center(50,"-"))
bowling_data["Bowling Figure"] = bowling_data.apply(lambda x : f'{x["Wickets"]}/{x["Given Runs"]}',axis=1)
print(bowling_data.head(),end="\n\n\n")
```

```python
#Overs Bowled: Total overs bowled.
print("Total overs bowled".center(50,"-"))
overs_bowled = pd.DataFrame(bowling_data.groupby("Player")["Overs"].sum()).reset_index()
overs_bowled = overs_bowled.rename(columns={"Overs":"Total Overs"})
print(overs_bowled.head(),end="\n\n\n")

#Average Balls: Average Balls bowled.

print("Average Balls bowled".center(50,"-"))
average_balls_bowled = pd.DataFrame(bowling_data.groupby("Player")["Balls"].mean()).reset_index()
average_balls_bowled["Balls"] = average_balls_bowled["Balls"].apply(lambda x : round(x,2))
average_balls_bowled = average_balls_bowled.rename(columns={"Balls":"Average Balls"})
print(average_balls_bowled.head(),end="\n\n\n")

#Maidens: Total maiden overs bowled (overs with no runs conceded).
```

```python
bowling_stats = total_wicket.merge(bowling_average).merge(economy_rate).merge(average_strike_rate).merge(five_wicket).merge(average_balls_bowled).merge(overs_bowled)
```
